```python
# How to add new record into existing csv-file...?
```

```python
# 1. Wrong-way --> adding new record to table

row = ['China','West','kim chan','11/19/2015 0:00:00','2305000','6870','21']
df.append(row)

# ERROR: **It has taken new-record in form of column(0)**
```

```python
# 2. Correct-way --> adding new record to table

row = ['China','West','kim chan','11/19/2015 0:00:00','2305000','6870','21']
```

```python
# pd.DataFrame helps to get an record in "SERIES"
# we mention "columns = df.columns.values" to match 'row'=record to respective column

pd.DataFrame([row] , columns = df.columns.values)
```

```python
# We named these pandas-process as "new_row" & append it in df

new_row = pd.DataFrame([row] , columns = df.columns.values)
```

```python
df.append(new_row)

# Append - Used to add new-record at bottom of df(table)
```
