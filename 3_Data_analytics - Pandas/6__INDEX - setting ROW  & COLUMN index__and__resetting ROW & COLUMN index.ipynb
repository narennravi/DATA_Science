```python
df = pd.read_csv("E:/DATA Science - My preparation/3_Data_analytics - Pandas/Data-files/Pandas - Q1 Sales.csv",header=0)
df.head()
```

```python
# COLUMN - index:-
```

```python
# 1. How to generate new-column...?
```

```python
po_number = [int(f"1000{i}")for i in range(20)]
```

```python
df['PO-Number'] = po_number
```

```python
# see the PO-Number column is added
df.head()
```

```python
# 2. how to address df based on PO-Number (it means set PO-Number as "PRIMARY-KEY")..?
```

```python
df.set_index('PO-Number').head()
```

```python
# 3. How to save changes to df which we done recently..?
#-----------------Inplace = True --------------------
```

```python
# Before saving changes
df.head()    
```

```python
po_number = [int(f"1000{i}")for i in range(20)]
df['PO-Number'] = po_number
```

```python
df.set_index('PO-Number', inplace=True)
```

```python
# After saving changes
df.head()
```
