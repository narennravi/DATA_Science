```python
#  3. How to groupby & want to address only specific column...?
#         * specific column - single
#         * specific column - mutiple
```

```python
df.groupby(['Country','Region'])['Total'].sum()

# Groupby coulumn: Country & Region 
# specific column:   Total - sum()
#  (single)
```

```python
df.groupby(['Country','Region','Sales Person'])['Total','Quantity'].sum()

# Groupby coulumn:   Country  &  Region 
# specific columns:  Total  &  Quantity - sum()
#    (multiple)
```

```python
# Dynamically:-
#      Without creating new column['Year'] we're supposed to address specific data from existing column 
      
```

```python
#  1. How to groupby 'Date of Purchase' & extract year-wise without creating year-column...?
#                        {----Dynamically----}
```

```python
#  see year-column is not here
df.head()
```

```python
# Solution:
df.groupby(df['Date of Purchase'].dt.year)['Quantity'].sum()
```

```python
# AGGREGATION - Function:-
```

```python
#        It is used to get set of multiple mathematical/statistical function in same line using --> .agg()
```

```python
#  1. How to use aggregation - .agg() in df to get multiple math-function...?
#           [Sum, Min, Max, Variance, Mean] = one line

#         * single (selective) column
#         * multiple (selective) columns
```

```python
df.head()
```

```python
# AGGREGATION - single (selective) column:-
```

```python
df.groupby(['Country','Region'])['Quantity'].agg(['sum','min','max','mean'])

# groupby  column         - 'Country' & 'Region'
# selective single column - 'Quantity'
# aggregation             - [sum,min,max,mean]
```

```python
# AGGREGATION - Multiple (selective) columns:-
```

```python
df.groupby(['Country','Region'])['Quantity','Selling Ratio'].agg(['sum','min','max','mean'])

# groupby  column         - 'Country' & 'Region'
# selective single column - 'Quantity' & 'Selling Ratio'
# aggregation             - [sum,min,max,mean]
```

```python
# Dynamically;-
#         Without separate 'year'-column.., we perform groupby based on year-wise by extracting year from
# 'Date of Purchase'-column using --> "".dt.year""

# This process is called Dynamic groupby
```

```python
# Similarly we can use AGGREGATION in Dynamic-Groupby also;-
```

```python
df.groupby(df['Date of Purchase'].dt.year)['Quantity'].agg(['sum','min'])
```
