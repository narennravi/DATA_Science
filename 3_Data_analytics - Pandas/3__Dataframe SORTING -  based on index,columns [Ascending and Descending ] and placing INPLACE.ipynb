```python
import pandas as pd
```

```python
df = pd.read_csv("E:/DATA Science - My preparation/3_Data_analytics - Pandas/Data-files/Pandas - Q1 Sales.csv")
df
```

```python
# IN-PLACE Function:-
```

```python
# 1. How to add new-index & save changes by using Inplace..?
```

```python
# We will be adding PO_Number as index to df
# So we are setting up index value(range) from 10000 onwards - random

po_num = [int(f"10000{i}") for i in range(20)]
import random
random.shuffle(po_num)
po_num

# Displays random 20-numbers from 10000(onwards)
```

```python
df["PO_Numbers"] = po_num
```

```python
# before inplace-function  
df.head()
```

```python
#Inplace = [True --->save changes to df]  ||  [False --->Don't save changes into df]

df.set_index("PO_Numbers", inplace=True)
```

```python
# After inplace-function
df.head()
```

```python
# SORTING - index:-
```

```python
# 1. How to sort index..?
# 1. How to sort index in Ascending..?
```

```python
df.sort_index()

# Displays sorted df in ascending-wise
```

```python
# NOTE:-  
#     1. Here PO_Number act as index
#     2. By default sorting process assigns df "Ascending order"
```

```python
# 2. How to sort index in descending order..?
```

```python
df.sort_index(ascending=False)

# Displays sorted df in Descending-wise(PO_Number/index)
```

```python
#  SORTING - Values:-
```

```python
#  1. How to sort-values based on the single - selective column..?
```

```python
df.sort_values('Country', ascending=False)

# Display df in **Descending order based on Country(column)**
```

```python
#  2. How to sort-values based on  Multiple - selected columns..?
```

```python
df.sort_values(["Country","Region"], ascending=False)

# Displays sorted values of "Country - descending"  &  "Region - descending"
```

```python
# INPLACE Function:-
```

```python
# NOTE:- 
#      if inplace=True ---> changes we done is preserved/saved in df
```

```python
#  BEFORE: inplace=True., {Unordered Sequences}
df.head()
```

```python
df.sort_values(['Country','Region'] , inplace=True)

# Sort_values based on column('Country' & 'Region')
```

```python
#  AFTER: inplace=True., {ordered Sequences}
df.head()
```
