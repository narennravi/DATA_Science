```python
import pandas as pd
```

```python
df = pd.read_csv("E:/DATA Science - My preparation/3_Data_analytics - Pandas/Data-files/Pandas - Q1 Sales.csv")
df.head()
```

```python
#  MERGE  &  JOIN:-
#       It can be performed only if we have 2-data-frames with different data and overlapping with same columns
#       we need new_df.,  Different-data.,  &  Overlaping columns 
```

```python
# NOTE:-
-    1. We already have default DataFrame   -->   df
-    2. Now we required another DataFrame   -->  new_df
```

```python
# Creating "new-DataFrame" --> new_df
```

```python
new_df = pd.DataFrame()

# Adding column for 'new_df' by referring existing "df" - columns
new_df['Employee Name'] = df['Sales Person'].unique()

import random
exp = [1,3,7]
random.shuffle(exp)

new_df['Experience in Company'] = exp
new_df['Total Experience'] = new_df['Experience in Company'] + random.randint(1,10)

# Displays "new_df"
new_df
```

```python
# NOTE:-
#    We added 3-columns @t "new-df" = [Employee Name , Exp in company , Total Exp]

**   Now we have 2-different data_frame - We can perform Merge/join   **
```

```python
------------------------------------------------------------------------------------------------------------------------------
```

```python
# MERGE:-
#       The data from "left"-dataframe  is merged with  "right"-dataframe

**    left  =  df.head()   
**    right =  new_df.head()
```

```python
# This is  Left - DataFRame
df.head() 
```

```python
# This is  Right - DataFrame
new_df
```

```python
# Merge - Function ---> [1st Attempt]:

df.merge(new_df)
```

```python
# REASON to solve error:-

#      We are supposed to mention either  left_on = None/Col-name.,  right_on = None/Col-name., 
#                                                                (or)  
#                                        lef_index = True/False.,   right-index = True/False.,

**MergeError:   No common columns to perform merge on. 
                Merge options: left_on=None, right_on=None, left_index=False, right_index=False **
```

```python
# Merge - Function ---> [2nd Attempt]:

df.merge(new_df,  left_on = 'Sales Person',  right_on = 'Employee Name').head(10)
```

```python
# 1. How to drop "Employee Name" from merge-function bcz both [Employee name & sales person] is repeated twice of name...?
               *****  Its our decision to drop any one-column --> among {Employee name/sales person}  ****
```

```python
df.merge(new_df,  left_on = 'Sales Person',  right_on = 'Employee Name').drop('Employee Name')

# Result as ERROR: **Bcz axis we didn't mention**
```

```python
df.merge(new_df,  left_on = 'Sales Person',  right_on = 'Employee Name').drop('Employee Name',axis=1)
```

```python
#  JOIN:-
#       The process is used to JOIN only "matching index-value[0,1,2]"  &  rest of other index remains "NaN" 
```

```python
df.head()
```

```python
new_df
```

```python
# JOIN - Function ----> [1st Attempt]:-

df.join(new_df).head(10)
```

```python
# NOTE:-
#       Its not fair while "join - 2 DataFrames[df & new_df]" & we get Nan for rest of column
```

```python
# REASON:-
#       Bcz join-function used to concentrate on "Index-value" of   [DataFrame = new_df]
#       new_df - index consist of only [0,1,2]. 
          -  only matching index[0,1,2] will "joined" & 
          -  rest of all index will be left "Nan"
```

```python
# JOIN - Function ----> [2nd Attempt]:-
```

```python
# 1. How to fix the join-function Correctly ...?
-       we need to set-index value for both DataFrames {df  &  new_df} 
-       And to add [inplace=True] function 
```

```python
df.set_index('Sales Person',  inplace = True)
new_df.set_index('Employee Name', inplace = True)
```

```python
df.head()
```

```python
new_df.head()
```

```python
# JOIN - Function ----> [2nd Attempt]:-

df.join(new_df)
```

```python
# NOTE:-
#    See Join & Merge is done Successfully 

# MERGE = It ask [left_on , right_on (or) left_index=True/false , right_index=True/false] for 2-DataFrames and it merges Perfectly
# JOIN  = We need to set_index for 2-DataFrames  &  it joins Perfectly
```
