```python
# Plots & graph:-
#     Plotting gives us Visualization of data in form of "graph/Chart" for easy understanding of data-frame for Analysing & Prediction


#NOTE:-
-     1. Plotting & graph representation "with groupby"-function -->  to avoid duplicates
-     2. Also we can perform "without groupby"-function          -->  Duplicate also plotted
```

```python
#  TYPES of GRAPH:-
-      1. Bar
-      2. line
-      3. Pie
-      4. Scatter

#  PROPERTIES:-
-      1. Title
-      2. Label
-      3. Color

#  SPECIAL Effect:-
-       1. Heatwaves - (Correlation-matrix)
```

```python
# Type-1:
# With Groupby-function:-
```

```python
# We assigned groupby_df for [Country] & selective col{Total & Quantity}
# grouby_df = while addressing for "graph-plot" it will be easy-access

groupby_df = df.groupby(['Country'])[['Quantity']].sum()
groupby_df
```

```python
# TYPES of Plotting:-

#    [First 4-Types were rapidly used in Data-Visualization..!]
#    [Rest of types were to display how to use..!]
```

```python
#  1. How to plot 'BAR'- graph for groupby_df..?
```

```python
groupby_df.plot.bar()
```

```python
#  2. How to plot 'line'- graph for groupby_df..?
```

```python
groupby_df.plot.line(color='Black')
```

```python
#  3. How to plot 'pie'- chart for groupby_df..?
```

```python
groupby_df.plot.pie(subplots=True)


# NOTE:-
#     For pie-chart mandatory to mention --> "subplots=True"
```

```python
#  4. How to plot 'Scatter'- graph for groupby_df..?
```

```python
groupby_df.plot.scatter(x=0, y='Quantity')


# NOTE:-
#    For Scatter-plot mandatory to mention ---> 'x=' & 'y='
```

```python
#  5. How to plot 'barh'- graph for groupby_df..?
```

```python
groupby_df.plot.barh(color='Purple')
```

```python
#  6. How to plot 'Area'- graph for groupby_df..?
```

```python
groupby_df.plot.area(color='Yellow')
```

```python
#  7. How to plot-'BOX' for groupby_df..?
```

```python
groupby_df.plot.box()
```

```python
#  8. How to plot 'denity'-graph for groupby_df..?
```

```python
groupby_df.plot.density(color='darkblue')
```

```python
#  9. How to plot 'hexbin'-graph for groupby_df..?
```

```python
groupby_df.plot.hexbin(x=0, y='Quantity')

# NOTE:-
#    For hexbin-plot mandatory to mention ---> 'x=' & 'y='
```

```python
#  10. How to plot 'hist'-graph for groupby_df..?
```

```python
groupby_df.plot.hist()
```

```python
# COLOR - graph:-

#       Its simpler by mentioning ---> (color='---name_of_color---')
```

```python
#   1. How to add "color" to graph-representation..?
```

```python
groupby_df.plot.barh(color='Palegreen')
```

```python
# LABEL - Graph:-

#    Some graph are required to mention the "x-label"  &  "y-label" value to generate the graphical-representation
#    Those graphs which need x & y are:-
             - Scatter plot  -->  (x='', y='')
             - Hexbin plot   -->  (x='', y='')
             - pie chart     -->  (subplots=True)
```

```python
# 1. SATTER plot  -->  (x='', y='')

groupby_df.plot.scatter(x=0, y='Quantity', title='Sales_QUANTITY of country', color='darkViolet')
```

```python
# 2. HEXBIN plot  -->  (x='', y='')

groupby_df.plot.hexbin(x=0, y='Quantity',title='Sales_QUANTITY of country', color='gold')
```

```python
# PIE Chart  -->  (subplots=True)

groupby_df.plot.pie(subplots=True)
```
