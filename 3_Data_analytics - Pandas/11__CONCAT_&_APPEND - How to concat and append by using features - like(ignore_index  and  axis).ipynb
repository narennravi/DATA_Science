```python
import pandas as pd
```

```python
df = pd.read_csv("E:/DATA Science - My preparation/3_Data_analytics - Pandas/Data-files/Pandas - Q1 Sales.csv")
df.head()
```

```python
# Conatenate  &  Merge:-
```

```python
# Concatenate:
#       Concat is "pandas"- method
#       It is used to combine mutiple Dataframes together as "one-data"
#       By pandas - we can have access (axis, index)

# Syntax :
        pd.concat([---df-names---])
```

```python
# we created multiple df's for understanding purpose:

new_df = df.copy()
new_df_1 = df.copy()
new_df_2 = df.copy()
```

```python
#  1.1  How to do concatenation using pandas [row-wise]...?
```

```python
pd.concat([new_df, new_df_1, new_df_2])
```

```python
#  1.2. How to ignore index-miscorrection during concat in pandas[row-wise]..?
```

```python
pd.concat([new_df, new_df_1, new_df_2], ignore_index=True)

# ignore_index = True  ---> It rearranges index in ordered sequences
```

```python
#  2.1  How to do concatenation using pandas [Column-wise]...?
```

```python
pd.concat([new_df, new_df_1, new_df_2], axis=1)
```

```python
# Append:-
#     Append is "df-method"  
#     It can append data only within (different datframes) i.e., from one-dataframe to another-dataframe
#     Here we *don't have* any access - (index, axis)

# Syntax:
       df.append([---df-names---])
```

```python
#  1. How to do append function in df..?
```

```python
df.append([new_df, new_df_1, new_df_2])
```

```python
# NOTE:-


# CONCAT - We can access/add both rows & columns using "Concatination" special features --> [axis, ignore_index]

# APPEND - We can only add data to record-rows & no special access
```
